2.Basic Arithmetic operations

Simple operations like addition,subtraction,multiplication and division

```python
#Addition
a=5;b=3
print("a+b=",a+b)
#Subtraction
print("a-b=",a-b)
#Multiplication
print("Multiplication:",a*b)
#Division
print("Division:",a/b)
```

3.**Variables and Data Types**

Indroduction to different data types likes integers , floats , strings and complex.

```python
#Integer
x=10
print("Data type of x is",type(x))
#float
y =3.14
print("Data type of y is",type(y))
#string
z="Python"
print("Data type of z is ",type(z))
#complex
c=2+3j
print("Data type of c is",type(c))
```

4. **Lists**

Creating and manipulating lists

```python
#Creating a list
fruits=["apple","banana","cherry"]
print("Lists:",fruits)
#Accessing elements
print("First fruit:",fruits[0])
#Adding an element
fruits.append("orange")
print("Updated list:",fruits)
#Removing an element
fruits.remove("banana")
print("list after removal:",fruits)
#Inserting an element at specific position in a list
fruits.insert(0,"banana")
print("New updated list:",fruits)
```

5. **Loops**

Using for and while loops.

```python
#for loop
for i in range(5):
  print("for loop iteration :",i)
```

```python
#while loop
count=0
while count<=5 :
  print("while loop iteration :",count)
  count += 1
```

6. **Conditional statement**

Using if, elif and else statements

```python
number=-5
if number > 0:
  print("Positive number")
elif number==0:
  print("Zero")
else:
  print("Negative number")
```

```python
a=input("Enter first number:")
b=input("Enter second number:")
if a > b:
  print(a,"is larger number")
elif a==b:
  print("Both are equal")
else:
  print(b,"is larger number")
```

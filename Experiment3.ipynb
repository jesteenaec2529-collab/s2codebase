##**Experiment No :3**

Evaluating gradient of a `scalar` field and visualize



**case** find the gradient of the scalar field $f=x^2+y^2+z^2$ and plot the gradient field.

##**Solution**

Evaluating the gradient of the scalar field

```python
import sympy as sp

#Define the symbols
x, y, z = sp.symbols('x y z')

#Define the scalar field
scalar_field =x**2+y**2+z**2  #Example scalar field
scalar_field

```

```python
# Compute the gradient

gradient =sp.Matrix([sp.diff(scalar_field,var) for var in(x,y,z)])
print("The gradient of the scalar field is")
gradient
```

##**Ploting the gradient of the scalar field**

```python
import numpy as np
import matplotlib.pyplot as plt

#Define a grid of points where you want to evaluate the gradient
x_vals = np.linspace(-10,10,20)
y_vals = np.linspace(-10,10,20)
z_vals = 0 # for 2D visulization in the z=0 plane

X,Y = np.meshgrid(x_vals,y_vals)

#Initialize arrays to store gradient components

u = np.zeros_like(X, dtype =float)
v = np.zeros_like(Y, dtype =float)

#Evauate the gradient at each point
for i in range(X.shape[0]):
  for j in range(X.shape[1]):
    grad_evaluated = gradient.subs({x:X[i,j],y:Y[i,j],z:z_val})
    u[i,j] = grad_evaluated[0]
    v[i,j] = grad_evaluated[1]


#Plot the gradient vectors using matplotlib
fig ,ax =plt.subplots()
ax.quiver(X,Y,u,v)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Gradient of the scalar field')

plt.show()
```
